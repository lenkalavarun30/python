```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = [c.lower() for c in s if c.isalnum()]
        return s == s[::-1]
```

```python
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def sod(n):
    d = [int(i) for i in str(n)]
    return sum(d)

def every(n):
    for i in range(3):
        d = n % 10
        n = n // 10
        if not prime(d):
            return False
    return True

for i in range(100, 1000):
    if prime(i) and prime(sod(i)) and every(i):
        print(i, end=" ")
```
